```python
plt.imshow(img)
# COLOR FORMAT
# MATPLOTLIB ----> RGB - RED GREEN BLUE
# OPEN CV ----> BGR - BLUE GREEN RED
```

```python
# Color Conversion & Correction
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# importing new image with color conversion 
img_gray = cv2.imread('E:/Languages/python/OpenCv/Computer-Vision-with-Python/DATA/00-puppy.jpg',cv2.
                      IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
# Resizing images with width and height
new_img = cv2.resize(fix_img,(1000,400))
plt.imshow(new_img)
```

```python
# Resizing images with ratios
w_ratio = 0.5
h_ratio = 0.5 
new_img = cv2.resize(fix_img,(0,0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
# flipping images 
new_img = cv2.flip(fix_img, 1) 
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
# flip(source , [-1 to 1 ]) horizontal to vertical
```

```python
cv2.imwrite('totally_new.jpg',new_img) # saving the file 
```

```python
fig = plt.figure(figsize=(10,8)) # blank canvas space 
ax = fig.add_subplot(111)
ax.imshow(new_img)
```
