```python
# Node 
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
# Linked list creatinon
class LinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node # pointing head to the new node
        self.tail = new_node # pointing tail to the new node 
        self.length = 1
        
# print the whole LinkedList 
    def print_list(self):
        temp = self.head # assigning the head of first node to temp variable
        while temp is not None: # having a while loop until the temp reaches to the end of linkedlist which 
            # is None
            print(temp.value) # printing the value of current node
            temp = temp.next # incrementing to the next node
# appending a New Node 
    def append(self, value): # appending means to add a new node at last
        new_node = Node(value) # creation of new node 
        # if there is no node at all (first-time creation)
        if self.length == 0:
            self.head = new_node # pointing head to the new node
            self.tail = new_node # pointing tail to the new node 
        # if already , there is more than one node 
        else:
            self.tail.next = new_node # pointing last node to new node 
            self.tail = new_node # pointing tail to new node 
        self.length += 1
        return True
# Popping an old Node 
    def pop(self):
        # if there is no node at all ,return none 
        if self.length == 0:
            return None
        # pointing both temp & pre to the head
        temp = self.head 
        pre = self.head
        while(temp.next): # 
            pre = temp
            temp = temp.next
        self.tail = pre
        self.tail.next = None
        self.length -= 1
        if self.length == 0:
            self.head = None
            self.tail = None
        return temp
    
    # prepend 
    def prepend(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.length += 1
        return True
#popping first node 
    def pop_first(self):
        #none condition
        if self.length == 0:
            return None
        temp = self.head 
        self.head = self.head.next 
        temp.next = None 
        self.length -= 1 
        # One node condition
        if self.length == 0:
            self.tail = None
        return temp
    
# get method 
    def get(self,index):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        for i in range(index):
            temp = temp.next
        return temp
# insert a node 
    def insert(self, index, value):
        if index < 0 or index > self.length:
            return False
        if index == 0:
            return self.prepend(value)
        if index == self.length:
            return self.append(value)
        new_node = Node(value)
        temp = self.get(index - 1)
        new_node.next = temp.next
        temp.next = new_node
        self.length += 1   
        return True  
# remove a node 
    def remove(self,index):
        # checking out of boundaries 
        if index < 0 or index >= self.length: 
            return None 
        # if index is first node 
        if index == 0: 
            return self.pop_first()
        # if indes is last node 
        if index == self.length -1: 
            return self.pop()
        # if index is in middle
        prev = self.get(index -1) # we use get() method to find the index and assign it to the prev
        temp = prev.next # assigning prev.next to the temp (temp == index)
        prev.next = temp.next # breaking the prev node , so that it can attach to temp.next node 
        temp.next = None # assigning the temp.next to None so that it will be removed
        self.length -= 1 # decrementing the length of linkedlist as a node is removed
        return temp 
```

```python
# when we write my_linked_list = LinkedList(4)
# it creates what up above we wrote with value of 4 to new_node
my_linked_list = LinkedList(0)
my_linked_list.append(2)

my_linked_list.insert(1,1)

my_linked_list.print_list()

```
